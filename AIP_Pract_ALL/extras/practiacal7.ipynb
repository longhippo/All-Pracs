#### Practical 7Write a program to apply various image enhancement using image derivatives by implementing smoothing, sharpening, and unsharp masking filters for generating suitable images for specific application requirements.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("lenna_original.jpg")
kernel_size = 5
```

```python
blur = cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)
laplacian = cv2.Laplacian(blur, cv2.CV_64F)
sharpened = img - laplacian
unsharp = cv2.addWeighted(img, 1.5, cv2.GaussianBlur(img, (kernel_size, kernel_size), 0), -0.5, 0)
```

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].imshow(img, cmap='gray')
axs[0, 0].set_title('Original Image')

axs[0, 1].imshow(laplacian, cmap='gray')
axs[0, 1].set_title('laplacian Blur')

axs[1, 0].imshow(sharpened, cmap='gray')
axs[1, 0].set_title('sharpned(img - l)')

axs[1, 1].imshow(unsharp, cmap='gray')
axs[1, 1].set_title('Unsharp Image')
for ax in axs.flat:
    ax.set_xticks([])
    ax.set_yticks([])
plt.show()
```
