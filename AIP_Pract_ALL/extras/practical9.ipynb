### Practical 9 Write the program to implement various morphological image processing techniques.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('lenna_original.jpg', 0)
kernel = np.ones((5, 5), np.uint8)
```

```python
dilation = cv2.dilate(img, kernel, iterations=1)
erosion = cv2.erode(img, kernel, iterations=1)

opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)
```

```python
fig, axs = plt.subplots(2, 4 ,figsize=(10, 5))

axs[0, 0].imshow(img, cmap='gray')
axs[0, 0].set_title('original Image')

axs[0, 1].imshow(dilation, cmap='gray')
axs[0, 1].set_title('dilated Image')

axs[0, 2].imshow(erosion, cmap='gray')
axs[0, 2].set_title('erosion Image')

axs[0, 3].imshow(opening, cmap='gray')
axs[0, 3].set_title('opening Image')

axs[1, 0].imshow(closing, cmap='gray')
axs[1, 0].set_title('closing Image')

axs[1, 1].imshow(gradient, cmap='gray')
axs[1, 1].set_title('gradient Image')

axs[1, 2].imshow(tophat, cmap='gray')
axs[1, 2].set_title('tophat Image')

axs[1, 3].imshow(blackhat, cmap='gray')
axs[1, 3].set_title('blackhat Image')

for ax in axs.flat:
    ax.set_xticks([])
    ax.set_yticks([])
fig.tight_layout()

plt.show()
```
